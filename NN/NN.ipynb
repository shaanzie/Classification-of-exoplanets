```python
dataset = pd.read_csv('new_data.csv')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
features=[]
for i in dataset.select_dtypes(include=['float64']):
    features.append(i)
features
```

```python
X = dataset.loc[:,features].values
```

```python
y = dataset.loc[:,['P. Zone Class']].values
```

```python
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(.95)
```

```python
principalComponents = pca.fit_transform(X)
```

```python
principalDf = pd.DataFrame(data = principalComponents)
```

```python
principalDf
```

```python
classes = dataset[['P. Zone Class','P. Mass Class', 'P. Composition Class', 'P. Atmosphere Class', 'P. Habitable Class']]
```

```python
classes
```

```python
for i in classes:
    principalDf[i] = dataset[i]
principalDf
```

```python
data_final = principalDf.values
```

```python
data_final
```

```python
X = data_final[:,0:21]
Y = data_final[:,21]
```

```python
from sklearn import preprocessing 
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
X_scale
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```python
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential([Dense(35,activation='sigmoid',input_shape=(21,)),
                    Dense(36,activation='sigmoid'),
                    Dense(37,activation='sigmoid'),
                    Dense(38,activation='sigmoid'),
                    Dense(39,activation='sigmoid'),
                    Dense(1,activation='sigmoid'),])
```

```python
from keras import optimizers
adam = optimizers.Adam(lr=0.01, beta_1=0.9, beta_2=0.990, amsgrad=False)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
hist=model.fit(X_train,Y_train,
               batch_size=2712,epochs=25)
```
